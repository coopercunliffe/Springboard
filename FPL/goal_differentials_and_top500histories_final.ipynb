```python
goals_conceded = teams[['id']]
goals_conceded = goals_conceded.set_index('id')
for gameweek in range(1, df['round'].max()+1):
    goals_conceded[gameweek] = 0
goals_conceded
```

```python
for gameweek in range(1, df['round'].max()+1):
    for opponent in range(1, 21):
        goals_conceded.loc[opponent, gameweek] = df[(df['round'] <= gameweek) & (df['opponent_team'] == opponent)].goals_scored.sum()

goals_conceded
```

```python
goals_conceded_rank = goals_conceded.rank(method='min')

goals_conceded_rank = goals_conceded_rank.astype(int)

# df['max_rank'] = df['Number_legs'].rank(method='max')
goals_conceded_rank
```

```python
def conceded_func(team, gameweek):
    return goals_conceded_rank.loc[team, gameweek]
```
