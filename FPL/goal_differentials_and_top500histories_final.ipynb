```python
df[(df['gameweek'] <= 18) & (df['opponent_team'] == 14) & (df['goals_scored'] >= 1)].sort_values('gameweek')[['gameweek', 'goals_scored', 'web_name']].goals_scored.sum()
```

```python
goals_conceded = teams[['id']]
goals_conceded = goals_conceded.set_index('id')
for gameweek in range(1, df['gameweek'].max()+1):
    goals_conceded[gameweek] = 0
goals_conceded
```

```python
for gameweek in range(1, df['gameweek'].max()+1):
    for opponent in range(1, 21):
        goals_conceded.loc[opponent, gameweek] = df[(df['gameweek'] <= gameweek) & (df['opponent_team'] == opponent)].goals_scored.sum()

goals_conceded
```

```python
def conceded_func(team, gameweek):
    return goals_conceded.loc[team, gameweek]
```

```python
conceded_func(15,3)
```

```python
team_week_pairs = []
for i in range(df.shape[0]):
    team_week_pairs.append((df.loc[i, 'opponent_team'], df.loc[i, 'gameweek']))
```

```python
df.head()
```

```python
df['team_a_conceded'] = [conceded_func(x, y) for (x, y) in team_week_pairs]
```
